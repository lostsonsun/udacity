---
**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
#求出利润，单位换算为百万
profit = (clean_md['revenue'] -  clean_md['budget'])/(10 ** 6)

plt.figure(figsize = [10, 5])

sb.violinplot(data = clean_md, x = 'release_year', y = profit, color = base_color, inner = None)
plt.xticks(rotation = 90)
plt.ylabel('profit (million)')
```

分析：
我先是用violin plot去分析电影净利润随年份变化的情况，但由于年份太多了，导致小提琴图被压缩的特别小，从小提琴图来看并不能得到很多有用的信息。
下面我用每年净利润的平均值来分析电影净利润与年份的关系。

```python
#求出每年净利润的平均值
average_revenue = clean_md.groupby(['release_year']).revenue.agg('mean')
average_budget = clean_md.groupby(['release_year']).budget.agg('mean')
average_profit = (average_revenue - average_budget) / (10 ** 6)

#画出净利润的平均值与年份的折线图
plt.figure(figsize = [10, 5])
sb.pointplot(x = average_profit.index, y = average_profit)
plt.xticks(rotation = 90)
plt.ylabel('profit (million)')
```

分析：
从上面的折线图可以看出，电影的净利润在1970年代到1980年代，电影票房平均净利润处在第一个高峰期。但是这段高峰期期间很不稳定，有大起大落。第二个高峰期是在1989到2004年期间，这段高峰期比第一个高峰期稳定，没有特别多的大起大落。最后从2005年开始到2015年，电影净利润在平均5000万到1亿左右稳定慢速地增长。
